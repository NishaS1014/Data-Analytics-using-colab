```python
import numpy as np

products = ["Highlighter", "Pen", "Water bottle", "Pencil", "Eraser"]
sales = np.array([42,100,67,89,67])
print("Total Sales:", np.sum(sales))        # 365
print("Average Sales:", np.mean(sales))     # 73.0
print("Highest Sales:", np.max(sales))      # 100
print("Lowest Sales:", np.min(sales))       # 42
best_index = np.argmax(sales)
print("Best Selling Product:", products[best_index])  # Pen
```
